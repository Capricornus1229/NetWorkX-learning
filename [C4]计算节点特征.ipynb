# 导入工具包

```python
# 图数据挖掘
import networkx as nx

# 数据可视化
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
%matplotlib inline

plt.rcParams['font.sans-serif']=['SimHei']  # 用来正常显示中文标签  
plt.rcParams['axes.unicode_minus']=False  # 用来正常显示负号
```

# 可视化辅助函数

```python
def draw(G,pos,measures,measures_name):
    nodes=nx.draw_networkx_nodes(G,pos,node_size=250,cmap=plt.cm.plasma,
                                 node_color=list(measures.values()),nodelist=measures.keys())
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01,linscale=1,base=10))
    edges=nx.draw_networkx_edges(G,pos)
    plt.title(measures_name)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
```

# 导入无向图

```python
G=nx.karate_club_graph()
pos=nx.spring_layout(G,seed=675)
nx.draw(G,pos=pos,with_labels=True)
```

# Jaccard Coefficient: 交并比

```python
preds=nx.jaccard_coefficient(G,[(0,4),(0,5),(0,6)])
for u,v,p in preds:
    print(f"({u},{v}) -> {p:.8f}")
```

# Adamic/Adar Index: Adamic/Adar指数

Adar指数用来衡量两个节点之间的相似性，定义为共同邻居节点的度的倒数之和，度越小的共同邻居节点对相似性的贡献越大。

```python
for u,v,p in nx.adamic_adar_index(G,[(0,4),(0,5),(0,6)]):
    print(f"({u},{v}) -> {p:.8f}")
```

# Katz Index: Katz指数

Katz指数通过计算两个节点之间所有可能路径的加权和来衡量它们的相似性，较短的路径权重更高。

```python
import numpy as np
import networkx as nx
from numpy.linalg import inv
G=nx.karate_club_graph()
```

```python
len(G.nodes)
```

```python
# 计算主特征向量
L=nx.normalized_laplacian_matrix(G)
e = np.linalg.eigvals(L.toarray())
print('最大特征值:',max(e))
# 折减系数
beta=1/max(e)
# 创建单位矩阵
I=np.identity(len(G.nodes))
# 计算Katz指数矩阵
Katz_matrix=inv(I - nx.to_numpy_array(G)*beta) - I
print(Katz_matrix.shape,Katz_matrix)
```
