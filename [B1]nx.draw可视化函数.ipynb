# 导入工具包

```python
import networkx as nx
import numpy as np
# 数据可视化
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False # 解决负号'-'显示为方块的问题
```

# 创建网格图

```python
G=nx.grid_2d_graph(4,4)  # 创建一个5x5的网格图
```

# 原生可视化

```python
pos=nx.spring_layout(G,seed=123)
nx.draw(G,pos)
```

# 不显示节点

```python
nx.draw(G,pos,node_size=0)
```

# 颜色设置

```python
# node_color：节点颜色 edgecolors：节点边框颜色 edge_color：边颜色 node_size：节点大小 with_labels：是否显示节点标签 width：边宽度
nx.draw(G,pos,node_color='red',edgecolors='blue',edge_color='green',node_size=100,with_labels=True,width=2)
```

# 有向图配置

```python
# tab:orange：橙色 tab:gray：灰色
nx.draw(G.to_directed(),pos,node_color='tab:orange',node_size=200,with_labels=False,edge_color='tab:gray',arrowsize=20,width=2)
```

# 设置每个节点坐标

```python
G=nx.Graph()
G.add_edges_from([(1,2),(1,3),(1,5),(2,3),(3,4),(4,5)])
nx.draw(G,with_labels=True)
```

```python
pos={1:(0,0),2:(-1,0.3),3:(2,0.17),4:(4,0.255),5:(5,0.03)}
options={
    'node_color':'skyblue',
    'node_size':500,
    'width':2,
    'edge_color':'gray',
    'with_labels':True,
    'font_size':16,
}
nx.draw(G,pos,**options)
ax=plt.gca() # 获取当前坐标轴
ax.margins(0.08) # 设置坐标轴边距
plt.axis("on") # 开启坐标轴显示
plt.grid(True) # 显示网格
plt.show()
```

```python
# 有向图同理
G=nx.DiGraph([(0,3),(1,3),(2,4),(3,5),(3,6),(4,6),(5,6)])
nx.draw(G,with_labels=True)
```

```python
left_nodes=[0,1,2]
mid_nodes=[3,4]
right_nodes=[5,6]
```

```python
# 可视化每个节点坐标
pos={n:(0,i) for i,n in enumerate(left_nodes)}
pos.update({n:(1,i+0.5) for i,n in enumerate(mid_nodes)})
pos.update({n:(2,i+0.5) for i,n in enumerate(right_nodes)})
pos
```

```python
nx.draw(G,pos,with_labels=True)
```

```python
# 节点也可以单独绘制
G=nx.house_graph()
nx.draw(G,with_labels=True)
```

```python
pos={0:(0,0),1:(1,0),2:(0,1),3:(1,1),4:(0.5,2.0)}
```

```python
plt.figure(figsize=(6,6))
# 绘制墙角四个节点
nx.draw_networkx_nodes(G,pos,nodelist=[0,1,2,3],node_color='lightgreen',node_size=500)
# 绘制屋顶节点
nx.draw_networkx_nodes(G,pos,nodelist=[4],node_color='orange',node_size=500)
# 绘制连接
nx.draw_networkx_edges(G,pos,alpha=0.5,width=6) #alpha:透明度
plt.axis('off')
plt.show()
```
