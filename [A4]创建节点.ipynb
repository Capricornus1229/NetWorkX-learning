# 导入工具包

```python
import networkx as nx

# 数据可视化
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False # 解决负号'-'显示为方块的问题
```

# 创建无节点，无连接空图

```python
G=nx.Graph()
G
```

```python
G.nodes
```

```python
nx.draw(G,with_labels=True)
```

# 添加单个节点

```python
G.add_node(1)
G.add_node('Tommy')
nx.draw(G,with_labels=True)
```

# 添加多个节点

```python
G.add_nodes_from([2,3,'Jerry','Spike'])
G.add_nodes_from(range(4,10))
nx.draw(G,with_labels=True)
```

# 添加带属性的节点

```python
G.add_nodes_from([
    (10,{"姓名":"Alice","年龄":25}),
    (11,{"姓名":"Bob","年龄":30}),
])
nx.draw(G,with_labels=True)
print(G.nodes[10]) # 输出节点10的属性
```

# 从别的图添加所有节点

```python
H=nx.path_graph(10)
nx.draw(H,with_labels=True)
```

```python
G.add_nodes_from(H)
nx.draw(G,with_labels=True)
G.nodes
```

# 将图本身作为【一个】节点添加

```python
G.add_node(H)
nx.draw(G,with_labels=True)
G.nodes
```
