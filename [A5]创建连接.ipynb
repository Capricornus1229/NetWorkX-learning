# 导入工具包

```python
import networkx as nx

# 数据可视化
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False # 解决负号'-'显示为方块的问题
```

# 创建图

```python
G=nx.Graph() # 创建一个无向图
print(G.is_directed())
H=nx.DiGraph() # 创建一个有向图
print(H.is_directed())
# 给整张图添加特征属性
G.graph['name']='my_graph'
print(G.graph)
```

# 创建节点

```python
G.add_node(0,feature='A',label=1,zihao=4) # 添加单个节点，并添加节点属性
G.add_nodes_from([
    (1,{'feature':'B','label':0,'zihao':3}),
    (2,{'feature':'C','label':1,'zihao':2}),
    (3,{'feature':'D','label':0,'zihao':1})
]) # 添加多个节点，并添加节点属性
```

# 全图节点信息

```python
G.number_of_nodes()  # 图中节点的数量
```

```python
G.nodes
```

```python
for node in G.nodes(data=True):
    print(node)
```

```python
nx.draw(G,with_labels=True)
```

# 创建连接

```python
# 创建单个连接
G.add_edge(0,1,weight=4.2) # 添加单个连接，并添加连接属性
nx.draw(G,with_labels=True)
```

```python
# 创建多个连接
G.add_edges_from([
    (1,2,{'weight':3.5}),
    (2,3,{'weight':2.8}),
    (3,0,{'weight':1.2})
]) # 添加多个连接，并添加连接属性
nx.draw(G,with_labels=True)
```

# 全图连接信息

```python
G.number_of_edges()  # 图中连接的数量
```

```python
G.edges(data=True) # 全图连接信息，并显示属性
```

```python
for edge in G.edges(data=True):
    print(edge)
```

# 查看指定节点连接数

```python
node_id=1
G.degree[node_id]  # 查看指定节点连接数
```

# 查看节点连接情况

```python
for neighbor in G.neighbors(node_id):
    print('Node{} is connected to Node{}'.format(node_id,neighbor))
```
