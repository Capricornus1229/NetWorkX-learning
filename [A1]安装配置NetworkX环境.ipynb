# 设置matplotlib中文显示

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.rcParams['font.sans-serif'] = ['SimHei'] # 设置中文字体为黑体
plt.plot([1,2,3],[100,500,300])
plt.xlabel('X轴标签',fontsize=14)
plt.ylabel('Y轴标签',fontsize=14)
plt.title('标题',fontsize=18)
plt.show()
```
